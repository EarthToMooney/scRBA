```python
import pandas as pd
```

```python
df_pro = pd.read_excel('./input/PROTEIN_stoich_curation_2021-05-21.xlsx')
df_pro.index = df_pro.id.to_list()
```

```python
df_aamap = pd.read_csv('./input/PROTEIN_amino_acid_map.txt', sep='\t')
df_aamap.index = df_aamap.aa_abbv.to_list()
```

```python
for p in df_pro.index:
    if pd.isnull(df_pro.loc[p, 'MW (g/mmol)']) == False:
        continue
    
    seq = df_pro.sequence[p]
    mw = 0
    for aa in df_aamap.index:
        mw += seq.count(aa) * df_aamap.MW[aa]
    mw = mw / 1000
    print(round(mw,5))
```
