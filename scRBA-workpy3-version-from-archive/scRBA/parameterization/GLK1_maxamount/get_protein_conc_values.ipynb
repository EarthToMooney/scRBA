```python
import pandas as pd
```

```python
df_dataidx = pd.read_csv('../input/data_list.txt', sep='\t')
```

```python
prot = 'YCL040W'
pvals = pd.core.series.Series(dtype='float64')
vtrans = pd.core.series.Series(dtype='float64')
for i in df_dataidx.index:
    data_id = df_dataidx.data_id[i]
    df_data = pd.read_excel('../datasets/' + data_id + '/' + data_id + '.xlsx')
    df_data.index = df_data.id.to_list()
    if prot in df_data.index:
        pvals[data_id] = df_data.loc[prot, 'conc (g/gDW)']
        vtrans[data_id] = df_data.loc[prot, 'vtrans (mmol/gDW/h)']
    else:
        pvals[data_id] = 0
        vtrans[data_id] = 0
```
