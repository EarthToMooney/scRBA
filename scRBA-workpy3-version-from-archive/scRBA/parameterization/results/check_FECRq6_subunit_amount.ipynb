```python
import pandas as pd
```

```python
df_enz = pd.read_excel('../../build_GAMS_model/input/ENZYME_stoich_curation.xlsx')
```

```python
df_enz[df_enz.rxn_src == 'FECRq6_m']
```

```python
text = df_enz.protein_stoich[3030]
subs = text.split(',')
```

```python
df_atps = pd.DataFrame(columns=['protein', 'stoich'])
for s in subs:
    p,v = s.split(':')
    df_atps.loc[p, 'protein'] = p
    df_atps.loc[p, 'stoich'] = int(v)
```

```python
datasets = [
    'Bjorkeroth2020_batchYNB', 'Elsemman2022_batchGlc',
    'Elsemman2022_chemoGlc020', 'Elsemman2022_chemoGlc023',
    'Elsemman2022_chemoGlc027', 'Elsemman2022_chemoGlc030',
    'Yu2021_chemoNlimNH4_005', 'Yu2021_chemoNlimNH4_010',
    'Yu2021_chemoNlimNH4_013', 'Yu2021_chemoNlimNH4_018',
    'Yu2021_chemoNlimNH4_030', 'Yu2021_chemoNlimNH4_035']

for dname in datasets:
    df_data = pd.read_excel('../datasets/' + dname + '/' + dname + '.xlsx')
    df_data.index = df_data.id.to_list()
    
    for p in df_atps.index:
        if p in df_data.index:
            df_atps.loc[p, 'MW (g/mmol)'] = df_data.loc[p, 'MW (g/mmol)']
            df_atps.loc[p, dname] = df_data.loc[p, 'conc (g/gDW)']
```

```python
df_atps.to_excel('./FECRq6_subunit_conc_g_per_gDW_frompy.xlsx', index=None)
```
