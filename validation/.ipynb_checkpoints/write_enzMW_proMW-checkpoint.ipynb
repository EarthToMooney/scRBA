```python
import pandas as pd
```

#### Enzyme MW

```python
df_enz = pd.read_excel('../build_GAMS_model/input/ENZYME_stoich_curation.xlsx')
```

```python
enz_mw = dict()
for i in df_enz.index:
    mw = df_enz.loc[i, 'MW (g/mmol)']
    if pd.isnull(mw) == False:
        enz_mw['ENZLOAD-' + df_enz.id[i][4:]] = mw
```

```python
enz_mw_text = []
for k,v in enz_mw.items():
    enz_mw_text.append(k + '\t' + str(v))
```

```python
with open('./enz_mw_g_per_mmol.txt', 'w') as f:
    f.write('\n'.join(enz_mw_text))
```

#### Protein MW

```python
df_pro = pd.read_excel('../build_GAMS_model/input/PROTEIN_stoich_curation.xlsx')
```

```python
pro_mw = dict()
for i in df_pro.index:
    mw = df_pro.loc[i, 'MW (g/mmol)']
    if pd.isnull(mw) == False:
        pro_mw[df_pro.id[i]] = mw
pro_mw['PROTDUMMY'] = 44.026303
```

```python
pro_mw_text = []
for k,v in pro_mw.items():
    pro_mw_text.append(k + '\t' + str(v))
```

```python
with open('./pro_mw_g_per_mmol.txt', 'w') as f:
    f.write('\n'.join(pro_mw_text))
```

```python
with open('./enz_mw_g_per_mmol.txt') as f:
    text = f.read().split('\n')
enz_mw = dict()
for i in text:
    k,v = i.split('\t')
    enz_mw[k] = float(v)
```

```python
with open('./pro_mw_g_per_mmol.txt') as f:
    text = f.read().split('\n')
pro_mw = dict()
for i in text:
    k,v = i.split('\t')
    pro_mw[k] = float(v)
```
