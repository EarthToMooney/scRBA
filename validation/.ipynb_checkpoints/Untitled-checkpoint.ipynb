```python
with open('../build_GAMS_model/model/RBA_rxns.txt') as f:
    text = f.read().split('\n')
text = [i for i in text if i not in ["", "/"]]
rxns = [i[1:-1] for i in text]
```

```python
biom_id = 'BIOSYN-BIODILAERO'
final_res = RBA_result(biom_id=biom_id);
```

```python
final_res.load_raw_flux(filepath='./runRBA.flux.txt')
```

```python
text_write = []
for rxn in rxns:
    if rxn in final_res.raw_flux.keys():
        v = final_res.raw_flux[rxn]
    else:
        v = 0
    text_write.append("'" + rxn + "' " + str(v))
text_write = ["/"] + text_write + ["/"]
```

```python
with open('./runRBA.vstart.txt', 'w') as f:
    f.write('\n'.join(text_write))
```
